```python
import numpy as np
import networkx as nx
import pickle as pkl
import matplotlib.pyplot as plt
import pandas as pd
from itertools import count
```

```python
with open('../dat/ind.credit.graph', 'rb') as f:
    full = pkl.load(f)
```

```python
train = pd.read_csv('../dat/clean.data')
```

## Subset by node range

```python
%%time

node_list = range(100)
subset = {subset:full[subset] for subset in node_list}

graph = nx.from_dict_of_lists(subset)
```

```python
print('number of edges:', len(graph.edges))
print('number of nodes:', len(graph.nodes))
```

```python
%%time

plt.figure(figsize=(15,15))
pos = nx.spring_layout(graph, k=0.1)

nc = nx.draw_networkx_nodes(graph, pos, node_size=50, alpha=0.2)
ec = nx.draw_networkx_edges(graph, pos, alpha=0.1)

plt.axis('off')
plt.show()
#plt.savefig('../img/graph.png', dpi=400)
```

## Subset by node range + edge filter

```python
%%time

# node filter
node_list = range(100)
subset = {subset:full[subset] for subset in node_list}

# edge filter (really a secondary node filter)
for i in node_list:
    subset[i] = [i for i in subset[i] if i < 10000]

graph = nx.from_dict_of_lists(subset)
```

```python
print('number of edges:', len(graph.edges))
print('number of nodes:', len(graph.nodes))
```

```python
%%time

plt.figure(figsize=(15,15))
pos = nx.spring_layout(graph, k=0.09)

nc = nx.draw_networkx_nodes(graph, pos, node_size=50, alpha=0.2)
ec = nx.draw_networkx_edges(graph, pos, alpha=0.1)

plt.axis('off')
plt.show()
#plt.savefig('../img/graph.png', dpi=400)
```

## colorize by label

```python
node_list = list(graph.nodes)

for i in range(len(list(graph.nodes))):
    graph.add_node(list(graph.nodes)[i], label=train['>50k $/yr'].iloc[i])
```

```python
%%time

plt.figure(figsize=(15,12))
pos = nx.spring_layout(graph, k=0.05)

groups = set(nx.get_node_attributes(graph, 'label').values())
mapping = dict(zip(sorted(groups),count()))
nodes = graph.nodes()
colors = [mapping[graph.node[n]['label']] for n in nodes]

nc = nx.draw_networkx_nodes(graph, pos, nodelist=nodes, node_color=colors, with_labels=False,
                            node_size=50, cmap='bwr', alpha=0.3)
ec = nx.draw_networkx_edges(graph, pos, alpha=0.1)

plt.colorbar(nc)
plt.axis('off')
plt.show()
#plt.savefig('../img/graph.png', dpi=400)
```
